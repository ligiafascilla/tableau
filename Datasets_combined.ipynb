```python
import pandas as pd
```

```python
#import May data
may = pd.read_csv("./202305-citibike-tripdata.csv")
may.head()
```

```python
#import June data
june = pd.read_csv("./202306-citibike-tripdata.csv")
june.head()
```

```python
#import July data
july = pd.read_csv("./202307-citibike-tripdata.csv")
july.head()
```

```python
#create dataset with combined data
combined_months = pd.concat([may, june, july], axis=0, ignore_index=True)
```

```python
combined_months.tail()
```

```python
#count rows
len(combined_months)
```

```python
#export new combined dataset as csv
combined_months.to_csv('combined_dataset.csv', index=False)
```
